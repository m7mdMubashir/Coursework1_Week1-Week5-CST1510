### Activity 3: Temperature Logger

Write a program that:
1. Asks the user for temperatures for 5 days
2. Writes each temperature to "temps.txt" with the day number
   Format: "Day 1: 22°C"
3. Reads the file and calculates the average temperature

<hr>

`````python
temperatures = []

# ask user for the temperature for 5 days
for i in range(1, 6):
    temp = int(input(f"Enter the temperature for day {i}: "))
    temperatures.append(temp)

# write each temperature to temps.txt
with open("temp.txt", "w") as file:
    for i, temp in enumerate(temperatures, start=1):
        file.write(f"Day {i}: {temp}°C\n")

# read file and calculate the average
total_temp = 0
count = 0

with open("temp.txt", "r") as file:
    for line in file:
        temp_str = line.strip().split(":")[1].strip().replace("°C", "")
        temp = int(temp_str)
        total_temp += temp
        count += 1

avg_temp = total_temp / count
print(f"The average temperature over {count} days is: {avg_temp:.2f}°C")
`````

### Activity 9: Finding "Dream" Sentences in Les_Brown.txt

From the `Les_Brown.txt` file, read the content and identify all sentences that contain the word "dream". Write these sentences to a new file called `dream_sentences.txt`.

**Steps:**
1. Open and read the content of `Les_Brown.txt`.
2. Split the text into sentences.
3. Filter out sentences that contain the word "dream" (case-insensitive).
4. Write these sentences into the new file `dream_sentences.txt`.
<hr>

`````python
with open("Les_Brown.txt", "r") as file:
    content = file.read()

# Split the text into sentences
import re
sentences = re.split(r'(?<=[.!?])\s+', content)

# Filter the sentences that contain the word "dream" (case-insensitive)
dream_sentences = [sentence for sentence in sentences if "dream" in sentence.lower()]

# new file with dream sentences
with open("dream_sentences.txt", "w") as output_file:
    for sentence in dream_sentences:
        output_file.write(sentence + "\n")
`````

### Activity 13: Identifying Duplicate Entries in Clothing Stock

Create a program that reads the `clothing_stock.txt` file and identifies if there are any duplicate entries (both item and size). Use a set to find and print these duplicates.

**Steps:**
1. Read the file and store each item in a set.
2. Keep track of duplicate entries.
3. Print any duplicates found.
<hr>

`````python
# open the file and read the lines
with open ("clothes.txt", "r") as file:
    items = [line.strip() for line in file if line.strip()]

# find duplicates using set
seen = set()
duplicates = set()

for item in items:
    if item in seen:
        duplicates.add(item)
    else:
        seen.add(item)

# print the duplicates if found any
if duplicates:
    print("duplicates: ")
    for d in duplicates:
        print(d)
else:
    print("no duplciates found")
`````

### Activity 14: Counting Different Sizes per Clothing Item

Create a Python program that reads data from the `clothing_stock.txt` file and counts how many different sizes of each clothing item are available. Store this in a dictionary and print it.

**Steps:**
1. Read the file and extract both clothing items and sizes.
2. Count how many different sizes exist for each item.
3. Store the results in a dictionary and print it.
<hr>

`````python
# read the file
with open ("clothes.txt", "r") as file:
    lines = [line.strip() for line in file if line.strip()]

#create dict to store items and their sized
item_sizes = {}

for line in lines:
    # split each line into item and sizes
    if " - " in line:
        item, size = line.split(" - ", 1)
        #add the size of item in set
        if item in item_sizes:
            item_sizes[item].add(size)
        else:
            item_sizes[item] = {size}

# count size and print
print("number of different sizes per clothing item: ")
for item, sizes in item_sizes.items():
    print(f"{item}: {len(sizes)} sizes ({", ".join(sorted(sizes))})")
`````

### Activity 15: Managing Study Groups

You are responsible for managing a study group for a course. Each student can be part of multiple study groups, and you need to track which groups a student is enrolled in. Students can join or leave study groups. What data structure would you choose to manage this efficiently, and why?

**Example data:**

```python
{"John": ["Math", "Physics"], "Emily": ["Physics", "Biology"], "Alex": ["Math"]}

Steps:

1. Choose a data structure that allows quick access to the groups each student belongs to.
2. Write a program that lets students join and leave study groups.
3. Implement a function to display all groups a student is part of.
<hr>

`````python
#set data
students = {
    "John": {"Math", "Physics"},
    "Emily": {"Physics", "Biology"},
    "Alex": {"Math"}
}

# func to add a student to a group
def join_group(student, group):
    if student not in students:
        students[student] = set() 
    students[student].add(group)
    print(f"{student} joined {group}")

# func to remove student from group
def leave_group(student, group):
    if student in students and group in students[student]:
        students[student].remove(group)
        print(f"{student} left the {group}")
    else:
        print(f"{student} is not in {group}")

# function to display all groups of a student
def show_groups(student):
    if student in students and students[student]:
        print(f"{student} is in groups: {', '.join(students[student])}")
    else:
        print(f"{student} is not in any group.")

join_group("Alex", "Physics")
leave_group("Emily", "Physics")
show_groups("John")
show_groups("Alex")
`````

### Activity 16: Managing and Shuffling Playlists

You are developing a music player that lets users create a playlist and shuffle it. Users should be able to add songs, remove songs, and shuffle the playlist randomly. What data structure would you use to represent the playlist, and why?

**Example data:**

```python
["Song A", "Song B", "Song C", "Song D"]
````

Steps:

1. Choose a data structure that allows easy addition, removal, and shuffling of songs.
2. Write a program that simulates playlist operations (add, remove, shuffle).
3. Let users shuffle the playlist and display the new order.
<hr>

`````python
import random  #to shuffle playlists


# use list 
playlist = ["Song A", "Song B", "Song C", "Song D"]

#func to add song 
def add_song(song):
    playlist.append(song)
    print(f"{song} added to the playlist")

# func to remove a song
def remove_song(song):
    if song in playlist:
        playlist.remove(song)
        print(f"{song} removed from playlist")
    else:
        print(f"{song} not found in the playlist")

# func to shuffle playlist
def playlist_shuffle():
    random.shuffle(playlist)
    print("playlist shuffled.")

# func to display the playlist
def show_playlist():
    print("Current playlist: ")
    for i, song in enumerate(playlist, 1):
        print(f"{i}. {song}")


#example
show_playlist()
add_song("Song E")
add_song("Song F")
remove_song("Song B")
playlist_shuffle()
show_playlist()
`````

### Activity 17: Counting Words in Les_Brown.txt

Read the content from `Les_Brown.txt`, and then count the number of occurrences of each word in the text. Write the word count results to a new file called `word_count.txt`.

**Steps:**
1. Open and read the contents of `Les_Brown.txt`.
2. Split the content into words and count the frequency of each word.
3. Write the word count in the format: `word: count` to a new file called `word_count.txt`.

<hr>

`````python
import string  

with open("Les_Brown.txt", "r") as file:
    text = file.read().lower()  # convert all to lowercase

# remove punctuation
for punctuation in string.punctuation:
    text = text.replace(punctuation, "")

# split text into words
words = text.split()

# count the word occurrences
word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# write the results to a file
with open("word_count.txt", "w") as other_file:
    for word, count in word_count.items(): 
        other_file.write(f"{word}: {count}\n")  # write each word and its count
`````

### Activity 19 (Set): Splitting Clothing Stock by Size

Create a Python program that reads the `clothing_stock.txt` file and splits the clothing items into separate sets based on their size (e.g., all XS items in one set, all M items in another set). Print each set.

**Steps:**
1. Read the file and separate items by their size.
2. Store each size in a different set.
3. Print each set of items.
<hr>

`````python
# open and read file
with open("clothes.txt", "r") as file:
    items = file.readlines()

# create a dictionary
size_sets = {}

# split items by size
for item in items:
    item = item.strip()  # remove newline character
    if "-" in item:
        name, size = item.split(" - ")
        if size not in size_sets:
            size_sets[size] = set() 
        size_sets[size].add(name)

# print each set
for size, items_set in size_sets.items():
    print(f"{size} items:", items_set)
`````
