4. **Body mass index (BMI)** is a measure of health and weight.

It can be calculated by taking your weight in kilograms and dividing it by the square of your height in meters. The interpretation of BMI for people 20 years or older is as follows:


BMI	Interpretation
BMI < 18.5	Underweight
18.5 ≤ BMI ≤ 25.0	Normal
25.0 ≤ BMI ≤ 30.0	Overweight
30.0 ≤ BMI 	Obese
 
Write a program that prompts the user to enter weight in pounds and height in inches and display the BMI. Note that one pound is 0.45359237 kilograms, and one inch is 0.0254 meters


````python
pounds = float(input("enter weight in pounds: "))
height = float(input("enter height in inches: "))
weight_in_kg = pounds * 0.45359237
height_in_meters = height * 0.0254

squared = height_in_meters ** 2

BMI = weight_in_kg / squared

print(f"Your BMI is {BMI:.2f}")
if BMI < 18.5:
    print("Underweight")
elif BMI <= 25.0:
    print("Normal")
elif BMI <= 30.0:
    print("Overweight")
else:
    print("Obese")
````

5. **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks ≥ 50) and failed.

````python
student_marks = [34, 51, 67, 86, 14, 99, 51, 13, 82]

total = 0 
for mark in student_marks:
    total += mark
average_mark = total / len(student_marks)

highest_mark = max(student_marks)
lowest_mark = min(student_marks)

passed = sum(1 for mark in student_marks if mark >= 50)
failed = sum(1 for mark in student_marks if mark < 50)

print(f"average mark of the class is {average_mark:.2f}")
print(f"highest mark achieved in the class is {highest_mark}")
print(f"the lowest mark achieved in the class is {lowest_mark}")
print(f"students who passed: {passed}")
print(f"students who have failed: {failed}")
````

12. **Dog Years**

It is commonly said that one human year is equivalent to 7 dog years. However, this simple conversion fails to recognise that dogs reach adulthood in approximately two years. As result, some people believe that it is better to count each of the first two human years as 10.5 dog years, and then count each additional human year as 4 dog years. 

Write a program that implements the conversion from human years to dog years described in the previous paragraph. Ensure that your program works correctly for conversions of less than two or more humans’ years. Your program should display an appropriate error message if the user enters a negative number. 


````python
hy = float(input("enter dog age in human years: "))

if hy <= 2:
    dy = hy * 10.5
else:
    dy = 2 * 10.5 + (hy - 2) * 4
print(f"the dog age in dog years will be: {dy}")
````

16. Write a **guessing game** in Python where the user has to guess a secret number. After every guess the program tells the user whether their number was too large or too small. At the end the number of tries needed should be printed.


````python
import random

secret_number = random.randint(1, 100)

tries = 0

print("you have to guess the number")

while True:
    guess = int(input("take a guess: "))
    tries += 1

    if guess < secret_number:
        print("too small")
    elif guess > secret_number:
        print("too big")
    else:
        print(f"congrats you guessed it in {tries} tries ")
        break
````

17. **Collecting Positive Numbers**
You are writing a small program to help a teacher keep track of positive numbers entered by students. The teacher also wants to know if any even numbers were given.

**Requirements:**

1. Keep asking the user to enter a positive number, or type `"done"` to stop.
2. Store all entered numbers in a **list**.
3. After the input ends:

   * If numbers were entered:

     * Print all the numbers entered.
     * Check the list and print the **first even number**. If no even number exists, print `"No even numbers found"`.
   * If no numbers were entered, print `"No numbers were entered"`.


> ### Note 
>
>* **`lower()`**: Converts any text into lowercase.
>  Example: `"DONE".lower()` becomes `"done"`.
>  This makes it easier to compare user input without worrying about uppercase/lowercase.
>
>* **`isdigit()`**: Checks if a string only contains digits (0–9).
>  Example: `"123".isdigit()` returns `True`, but `"12a".isdigit()` returns `False`.
>  This helps us make sure the user typed a valid number before converting it into an integer.
> 


---

````python

positive_numbers = []

while True:
    userinput = input("enter a positive number or type 'done' to finish: ").lower()

    if userinput == "done":
        break

    if userinput.isdigit():
        number = int(userinput)
        if number > 0:
            positive_numbers.append(number)
        else:
            print("enter a positive number.")
    else:
        print("enter a positive number or 'done'.")

if positive_numbers:
    print("\nNumbers entered:", positive_numbers)

    found_even = False
    for num in positive_numbers:
        if num % 2 == 0:
            print("first even number:", num)
            found_even = True
            break

    if not found_even:
        print("no even numbers found.")
else:
    print("no numbers were entered.")
````

18. Suppose you want to develop a **program to play a lottery.**
    
The program randomly generates a two-digit number, prompts the user to enter a <u>two-digit number</u>, and determines whether the user wins according to the following rules:


    1. If the user’s input matches the lottery in the exact order, the award is £10,000.
    2. If all the digits in the user’s input match all the digits in the lottery number, the award is £3,000.
    3. If one digit in the user’s input matches a digit in the lottery number, the award is £1,000.

21. **Password cracker** - write a program to print the accuracy of password guess.

For example: The actual password is `MDX123` while a user tried `MDX321` or `mdx123`, the program should tell that how accurately the password was guessed, in this case it should be `50%` accuracy. if a user enter `MDX124`, the output should be `83%` and in case of `123MDX` the output should be `0%`.

Points to keep in mind while calculating the accuracy a) case of the letters entered b) order of the letters

If a password is accurately guessed, the output should state `Password cracked!`

````python
actual_password = 'MDX123'

guess = input("guess the password; ")
if len(guess) != len(actual_password):
    print("incorrect. Accuracy 0%")
else:
    match = 0
    for i in range(len(actual_password)):
        if guess[i] == actual_password[i]:
            match += 1

    accuracy = (match / len(actual_password)) * 100
    if accuracy == 100:
        print("password cracked!")
    else:
        print(f"Accuracy: {accuracy:.0f}")
````
